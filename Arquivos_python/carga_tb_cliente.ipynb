```python
import pandas as pd
import pyodbc

server = 'VENDAS' # substitua pelo nome do seu servidor 
database = 'Python' # substitua pelo nome do seu Banco de dados
conexaoDB = pyodbc.connect('DRIVER={ODBC DRIVER 17 for SQL Server};'
                           f'SERVER={server};'
                           f'DATABASE={database};'
                           'Trusted_Connection=yes;')
                           
                           
                           
cursor = conexaoDB.cursor() #criando cursor de comando
```

```python
dados = pd.read_csv(r"C:\Users\supor\Downloads\SQL_Server\arquivos_csv\Clientes.csv" , delimiter= ',')
dados.head(5)
```

```python
for coluna in dados.columns:
    print(f'coluna: {coluna}, tipo de dados: {dados[coluna].dtypes}')
```

```python
dados['number'] = dados['number'].astype(str)
```

```python
dados['created_at'] = pd.to_datetime(dados['created_at'])
dados['email'] = dados['email'].fillna('Sem registro')
dados['street'] = dados['street'].fillna('Sem registro')
dados['number'] = dados['number'].fillna('Sem registro')
dados['additionals'] = dados['additionals'].fillna('Sem registro')
dados['state'] = dados['state'].fillna('Sem registro')
```

```python
dados.head(5)
```

```python
str(dados.columns).replace("'","")# coluna de origem 
```

```python
cursor.execute('truncate table [clientes]') # executa a tarefa de apagar dados 
cursor.commit()
```

```python
for coluna in dados.columns:
    print(f'coluna: {coluna}, tipo de dados: {dados[coluna].dtypes}')
```

```python
# INSERÇÃO USANDO TUPLA DE PARAMETROS
for index,linha in dados.iterrows():
    linha.email = str(linha.email) # converter para o tipo (str)
    linha.state = str(linha.state)
    linha.street = str(linha.street)
    linha.number = str(linha.number)
    linha.additionals = str(linha.additionals)
    
    cursor.execute("insert into[clientes](id,created_at,first_name,last_name,email,cell_phone,country,state,street,number,additionals)values(?,?,?,?,?,?,?,?,?,?,?)"  
                   ,linha.id,linha.created_at,linha.first_name,linha.last_name,linha.email,linha.cell_phone,linha.country,  
                   linha.state,linha.street,linha.number,linha.additionals)
    
cursor.commit() # validar dados no SQL server
cursor.close() # Fechar cursor 
conexaoDB.close() #Fechar conexão
```
