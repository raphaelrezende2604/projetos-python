```python
import pandas as pd
import pyodbc

server = 'VENDAS' # substitua pelo nome do seu servidor 
database = 'Python' # substitua pelo nome do seu Banco de dados
conexaoDB = pyodbc.connect('DRIVER={ODBC DRIVER 17 for SQL Server};'
                           f'SERVER={server};'
                           f'DATABASE={database};'
                           'Trusted_Connection=yes;')
                           
                           
                           
cursor = conexaoDB.cursor() #criando cursor de comando
```

```python
query = ''' select 
 t1.id as idVenda, t1.order_id,t1.total_price,
 t2.created_at, t2.customer_id, t2.Status_,t3.NOME as Produto, t4.NOME as categoria,
 t5.last_name,t5.first_name,t5.cell_phone,t5.email,t5.country,t5.state
 


 from items as t1
 left join ordens as t2 on t1.order_id=t2.id
 left join produto as t3 on t1.product_id = t3.ID
 left join categoria as t4 on t4.ID = t3.ID_CATEGORY
 left join clientes as t5 on t5.id = t2.customer_id'''

resultadoQuery = pd.read_sql(query,conexaoDB)
resultadoQuery.head(20)
```
