```python
import pandas as pd
df = pd.read_csv(r"C:\Users\supor\Downloads\SQL_Server\arquivos_csv\Clientes.csv" , delimiter= ',')
df.head(20)
```

```python
# Converter a coluna 'created_at' para o formato data
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
# juntando nome e sobrenome
df['Nome completo'] = df['first_name'] + ' '+ df['last_name']
df.head(10)
```

```python
# Adicionar coluna com o Ano
df['ano'] = df['created_at'].dt.year
df.head(15)
```

```python
# Converter a coluna 'created_at' para o formato data
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
df['mes'] = df['created_at'].dt.month # Adicionar coluna com Mês
df.head(10)
```

```python
df['Dia'] = df['created_at'].dt.day
df.head(10)
```

```python
# Adicionar coluna  com o nome do mês  em português
meses = ['Janeiro','Fevereiro','Março', 'Abril','Maio','Junho',
         'Julho','Agosto','Setembro','Outubro','Novembro', 'Dezembro']

df['Nomemes'] = df['created_at'].dt.month.map(lambda x: meses[x-1])
df.head(10)
```

```python
df['Hora'] = df['created_at'].dt.hour
df['minuto'] = df['created_at'].dt.minute
df['segundo'] = df['created_at'].dt.second
df.head(10)
```
